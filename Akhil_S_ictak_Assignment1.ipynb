```python

plt.scatter(data['SL'], data['Classification'],color='red', alpha=0.5)
plt.scatter(data['PL'], data['Classification'],color='blue', alpha=0.8)
plt.scatter(data['SW'], data['Classification'],color='green', alpha=0.7)
plt.scatter(data['PW'], data['Classification'],color='black', alpha = 0.6)
plt.xlabel('SL - PL - SW - PW')
plt.show()
```
